```python
import numpy as np

def politica_inventario(K, d, h, c1, c2, q):
    # 1. Calculamos la Q optima (EOQ) basica
    Qop = np.sqrt((2*K*d)/h)

    # 2. Revisamos si Qop ya cae en la zona de descuento
    if Qop >= q:
        return Qop, "Zona II (EOQ con descuento)"

    # 3. Si no, tenemos que resolver la ecuación cuadrática
    # Primero calculamos el Costo Total en Qop con el precio alto (c1)
    TC_Qop = c1*d + (K*d)/Qop + (h*Qop)/2

    # Usamos los coeficientes de la ecuación que vimos en el ejemplo (In 12)
    # La ecuación es: Q^2 + bQ + c = 0
    b = (2/h) * (c2*d - TC_Qop)
    c = (2*K*d)/h

    # Resolvemos con la fórmula general (-b +/- raiz(b^2 - 4ac)) / 2a
    discriminante = b**2 - 4*c

    # Obtenemos la solución 's' (la raiz positiva mayor)
    s = (-b + np.sqrt(discriminante)) / 2

    # 4. Seleccionamos la solución correcta comparando q y s
    # La condición es: Qop < q < s
    if q < s:
        return q, "Zona II (Se pide q para aprovechar descuento)"
    else:
        return Qop, "Zona I (No conviene el descuento)"

# --- Probamos la funcion con los datos del ejemplo ---
c1 = 3
c2 = 2.5
d = 187.5
h = 0.02
K = 20
q = 1000

cantidad, zona = politica_inventario(K, d, h, c1, c2, q)
print(f"Cantidad a pedir: {cantidad}")
print(f"Region: {zona}")
```

```python
import numpy as np

def obtener_politica_optima(K, d, h, c1, c2, q, L):
    """
    Calcula la política óptima de inventario (Q* y ROP).
    Devuelve un mensaje con la instrucción precisa.
    """

    # --- PASO 1: Calcular la Cantidad a Pedir (Q*) ---
    # Calculamos EOQ base
    Qop = np.sqrt((2*K*d)/h)

    # Lógica de decisión de precios (igual que Ejercicio 1)
    if Qop >= q:
        Q_star = Qop
    else:
        # Resolvemos la ecuación cuadrática interna
        TC_Qop = c1*d + (K*d)/Qop + (h*Qop)/2
        b = (2/h) * (c2*d - TC_Qop)
        c = (2*K*d)/h

        discriminante = b**2 - 4*c
        if discriminante >= 0:
            s = (-b + np.sqrt(discriminante)) / 2

            # Decisión final comparando q y s
            if q < s:
                Q_star = q  # Conviene el descuento
            else:
                Q_star = Qop # No conviene
        else:
            Q_star = Qop

    # --- PASO 2: Calcular el Punto de Reorden (ROP) ---
    # Calculamos el tiempo de ciclo (cuánto dura el inventario Q*)
    t1 = Q_star / d

    # Ajuste por si el tiempo de entrega (L) es mayor que el ciclo
    # (Lógica vista en tu captura 'In [20]')
    n_ciclos = np.floor(L / t1)
    Le = L - (n_ciclos * t1)  # Tiempo de entrega efectivo

    ROP = Le * d  # Inventario al cual debemos reordenar

    return f"Pedir {Q_star:.2f} cuando el inventario caiga a: {ROP:.2f}"

# --- PROBAR LA FUNCIÓN CON TUS DATOS ---
c1 = 3
c2 = 2.5
d = 187.5
h = 0.02
K = 20
L = 2
q = 1000

mensaje = obtener_politica_optima(K, d, h, c1, c2, q, L)
print(mensaje)
```
